# REINFORCE Algo
aka Monte Carlo Policy Gradient: update params at every step instead of only after N steps

https://www.youtube.com/watch?v=5eSh5F8gjWU&ab_channel=AndriyDrozdyuk

The simplest policy gradient algorithm is the REINFORCE algorithm, also known as Monte Carlo Policy Gradient. It is a basic and fundamental policy gradient method that is easy to understand and implement.

The term "REINFORCE" is often written in all capital letters to emphasize its origin as an acronym. REINFORCE stands for "REward Increment = Nonnegative Factor × Offset Reinforcement × Characteristic Eligibility." It was introduced in the seminal paper "Simple Statistical Gradient-Following Algorithms for Connectionist Reinforcement Learning" by Ronald J. Williams in 1992.

```python
# quick test to see how the prob of action comes into play

probs=torch.tensor([0.1,0.9])
prob_dist=torch.distributions.Categorical(probs)
n_0,n_1=0,0
for _ in range(10):
    action=prob_dist.sample()
    if action==torch.tensor(0):
        n_0+=1
    else:
        n_1+=1
    print(action) #0.1 is action 0, 0.9 is action 0
    
print()
print(f"num times action 0 was selected: {n_0}")
print(f"num times action 1 was selected: {n_1}")
#this shows how the sampling depends on the prob distribution and is NOT at random
```

## implementation

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import gym
```

```python
lr=0.001
gamma=0.95

env=gym.make("CartPole-v0")

model=nn.Sequential(
    nn.Linear(4,64), #4 is the input of states
    nn.ReLU(),
    nn.Linear(64,env.action_space.n),
    nn.Softmax(dim=-1)
)

optimizer=optim.Adam(model.parameters(),lr=lr)
```

```python
# gather 1 episode of experiences - this is 1 trajectory

state=torch.tensor(env.reset(),dtype=torch.float)
done=False
Actions,States,Rewards=[],[],[]

while not done:
    action_probs=model(state)
    prob_dist=torch.distributions.Categorical(probs=probs)
    action=prob_dist.sample().item()
    new_state,reward,done,_=env.step(action)
    
    Actions.append(torch.tensor(action,dtype=torch.int))
    States.append(state)
    Rewards.append(reward)
    
    state=torch.tensor(new_state,dtype=torch.float)
    
# loop thru all timesteps in an episode and update weights (this is the Monte Carlo method - updating params at every step)
DiscountedReturns=[]
for step in range(len(Rewards)): 
    G=0.0 #this is to calc G0,G1,G2,G3,G4,...
    for i,rew in enumerate(Rewards[step:]):
        G+=(gamma**i)*rew
    DiscountedReturns.append(G)
```

```python
# iterate thru episode variables again and update policy

for State,Action,G in zip(States,Actions,DiscountedReturns):
    action_probs=model(State)
    prob_dist=torch.distributions.Categorical(action_probs)
    log_prob=prob_dist.log_prob(Action)
    
    loss= - log_prob*G #minus sign to reverse direction of gradient descent to gradient ascent
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

# full code

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import gym
import time

LR=0.001
GAMMA=0.99
MAX_STEPS=200
N_EPISODES=1000
PRINT_STEPS=200

env=gym.make("CartPole-v0")

model=nn.Sequential(
    nn.Linear(4,64), #4 is the input of states
    nn.ReLU(),
    nn.Linear(64,env.action_space.n),
    nn.Softmax(dim=-1)
)

optimizer=optim.Adam(model.parameters(),lr=LR)

start_time=time.time()
# gather 1 episode of experiences - this is 1 trajectory
All_rewards=[]
for eps in range(N_EPISODES):
    state=torch.tensor(env.reset(),dtype=torch.float)
    done=False
    Actions,States,Rewards=[],[],[]

    for step in range(MAX_STEPS):
        action_probs=model(state)
        prob_dist=torch.distributions.Categorical(probs=probs)
        action=prob_dist.sample().item()
        new_state,reward,done,_=env.step(action)

        Actions.append(torch.tensor(action,dtype=torch.int))
        States.append(state)
        Rewards.append(reward)

        state=torch.tensor(new_state,dtype=torch.float)

    # loop thru all timesteps in an episode and update weights (this is the Monte Carlo method - updating params at every step)
    DiscountedReturns=[]
    for step in range(len(Rewards)): 
        G=0.0 #this is to calc G0,G1,G2,G3,G4,...
        for i,rew in enumerate(Rewards[step:]):
            G+=(gamma**i)*rew
        DiscountedReturns.append(G)
        
    # iterate thru episode variables again and update policy

    for State,Action,G in zip(States,Actions,DiscountedReturns):
        action_probs=model(State)
        prob_dist=torch.distributions.Categorical(action_probs)
        log_prob=prob_dist.log_prob(Action)

        loss= - log_prob*G #minus sign to reverse direction of gradient descent to gradient ascent

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    All_rewards.append(np.mean(Rewards))
        
    if (eps+1)%PRINT_STEPS==0:
        ave_rew=np.mean(All_rewards[-200:])
        print(f"Average rewards past {PRINT_STEPS} episodes: {ave_rew}, Time elapsed: {time.time()-start_time}")
        
print()
print("############# END TRAINING #############")
print(f"Total time taken: {time.time()-start_time}")
```

```python
# test the env

for _ in range(10):
    state=torch.tensor(env.reset(),dtype=torch.float)
    done=False
    env.render()
    while not done:
        p=model(state)
        c=torch.distributions.Categorical(probs=p)
        action=c.sample().item()

        new_state,reward,done,_=env.step(action)
        env.render()

        state=torch.tensor(new_state,dtype=torch.float)
```
