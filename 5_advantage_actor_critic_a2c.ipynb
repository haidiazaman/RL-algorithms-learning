state value function, Vpi(s): 
The state-value function for policy pi, denoted as Vpi, tells us how good any given state is for an agent following policy. In other words, it gives us the value of a state under policy pi.
Formally, the value of state s under policy pi is the expected return from starting from state s at time t and following policy pi thereafter. 


    
action value function, Qpi(s,a):
Similarly, the action-value function for policy pi, denoted as Qpi, tells us how good it is for the agent to take any given action from a given state while following policy pi. In other words, it gives us the value of an action under policy pi.
Formally, the value of action a, in state s under policy pi is the expected return from starting from state s at time t, taking action a, and following policy pi thereafter. 

algo has 2 models

critic: estimates the value function. this is usually the action-value (q value) [can be state-value as well]
actor: updates the policy distribution in the direction suggested by the critic 
    
Advantage: using the V function as the baseline function, we subtract the Q value term with the V value. Intuitively, this means how much better it is to take a specific action compared to the average, general action at the given state. We will call this value the advantage value:

A(s,a)=Q(s,a)-V(s) [Q is the action value] -----(1)
Bellman optimality equation: Q(s,a)=E[rt+1 + gamma*V(st+1)] ------(2)
sub 2 into 1:
A(s,a) = rt+1 + gamma*V(st+1) - V(s) [E expecation term dissapears - not sure why]

This model is the Advantage Actor Critic (A2C)

On each learning step, we update both the 
Actor parameter (with policy gradients and advantage value), and 
the Critic parameter (with minimizing the mean squared error with the Bellman update equation)

Entropy regularization encourages the policy (actor) to explore more diverse actions by penalizing low-entropy policies. Entropy, in this context, refers to the uncertainty or randomness of the policy's action distribution. By adding an entropy term to the objective function of the actor, the algorithm is incentivized to take actions with higher entropy, i.e., actions with more exploration and less deterministic behavior.

When the entropy regularization coefficient is increased, the algorithm becomes more sensitive to the entropy term, resulting in a higher emphasis on exploration. The increased entropy encourages the actor to take actions with higher diversity, exploring different possibilities and reducing the chances of getting stuck in suboptimal solutions.

By increasing exploration, the actor-critic algorithm can discover better policies and potentially find higher-rewarding trajectories. This exploration-exploitation balance is crucial in reinforcement learning, as it allows the algorithm to trade off between exploring new actions and exploiting the current knowledge to maximize long-term rewards.1`

# algo 

```python
import sys
import torch
import gym
import time
import numpy as np
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable
import matplotlib.pyplot as plt
import pandas as pd
```

```python
HIDDEN_SIZE=256
LR=3E-4
GAMMA=0.99 #DECAY CONSTANT
BETA=0.001 #ENTROPY REGULARIZATION TERM - increase this value to encourage more exploration
MAX_STEPS=300
N_EPS=3000
```

```python
class ActorCritic(nn.Module):
    def __init__(self,n_inputs,n_actions,hidden_size,lr=3e-4):
        super(ActorCritic,self).__init__() 
        
        self.critic_linear1=nn.Linear(n_inputs,hidden_size)
        self.critic_linear2=nn.Linear(hidden_size,1)        
        
        self.actor_linear1=nn.Linear(n_inputs,hidden_size)
        self.actor_linear2=nn.Linear(hidden_size,n_actions)        
        
    def forward(self,state):
        state=Variable(torch.from_numpy(state).float().unsqueeze(0))
        
        value=F.relu(self.critic_linear1(state))
        value=self.critic_linear2(value)
        
        policy_dist=F.relu(self.actor_linear1(state))
        policy_dist=F.softmax(self.actor_linear2(policy_dist),dim=1)        
        
        return value,policy_dist
```

```python
def a2c(env):
    start_time=time.time()
    num_inputs = env.observation_space.shape[0]
    num_outputs = env.action_space.n
    
    actor_critic = ActorCritic(num_inputs, num_outputs, HIDDEN_SIZE)
    ac_optimizer = optim.Adam(actor_critic.parameters(), lr=LR)

    all_lengths = []
    average_lengths = []
    all_rewards = []
    entropy_term = 0

    for episode in range(N_EPS):
        log_probs = []
        values = []
        rewards = []

        state = env.reset()
        for steps in range(MAX_STEPS):
            value, policy_dist = actor_critic.forward(state)
            value = value.detach().numpy()[0,0]
            dist = policy_dist.detach().numpy() 

            action = np.random.choice(num_outputs, p=np.squeeze(dist))
            log_prob = torch.log(policy_dist.squeeze(0)[action])
            entropy = -np.sum(np.mean(dist) * np.log(dist))
            new_state, reward, done, _ = env.step(action)

            rewards.append(reward)
            values.append(value)
            log_probs.append(log_prob)
            entropy_term += entropy
            state = new_state
            
            if done or steps == MAX_STEPS-1:
                Qval, _ = actor_critic.forward(new_state)
                Qval = Qval.detach().numpy()[0,0]
                all_rewards.append(np.sum(rewards))
                all_lengths.append(steps)
                average_lengths.append(np.mean(all_lengths[-10:]))
                time_taken=round(time.time()-start_time,1)
                if episode % 250 == 0 and episode!=0:                    
                    sys.stdout.write("episode: {}, episode reward: {}, steps taken: {}, time elapsed: {}s \n".format(episode, np.sum(rewards), steps,time_taken))
                break
        
        # compute Q values
        Qvals = np.zeros_like(values)
        for t in reversed(range(len(rewards))):
            Qval = rewards[t] + GAMMA * Qval
            Qvals[t] = Qval
  
        #update actor critic
        values = torch.FloatTensor(values)
        Qvals = torch.FloatTensor(Qvals)
        log_probs = torch.stack(log_probs)
        
        advantage = Qvals - values
        actor_loss = (-log_probs * advantage).mean()
        critic_loss = 0.5 * advantage.pow(2).mean()
        ac_loss = actor_loss + critic_loss + BETA * entropy_term

        ac_optimizer.zero_grad()
        ac_loss.backward()
        ac_optimizer.step()

        
    
    # Plot results
    smoothed_rewards = pd.Series.rolling(pd.Series(all_rewards), 10).mean()
    smoothed_rewards = [elem for elem in smoothed_rewards]
    plt.plot(all_rewards)
    plt.plot(smoothed_rewards)
    plt.plot()
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.show()

    plt.plot(all_lengths)
    plt.plot(average_lengths)
    plt.xlabel('Episode')
    plt.ylabel('Episode length')
    plt.show()
```

```python
if __name__ == "__main__":
    env = gym.make("CartPole-v0")
    a2c(env)    
```
